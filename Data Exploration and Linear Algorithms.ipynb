Data Preparation

```python
dataset_test= pd.read_csv('C:\\hudsondata\\Machine Learning\\Kaggle\\AllState\\test.csv')
```

```python
dataset_test = dataset_test.iloc[:,1:]
```

```python
labels = []
for i in range(0, 116):
    train = dataset[cols[i]].unique()
    test = dataset_test[cols[i]].unique()
    labels.append(list(set(train)|set(test)))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
cats =[]
for i in range(0, 116):
    #Label Encode
    label_encoder = LabelEncoder()
    label_encoder.fit(labels[i])
    feature = label_encoder.transform(dataset.iloc[:,i])
    feature = feature.reshape(dataset.shape[0],1)
    #One Hot Encode
    onehot_encoder = OneHotEncoder(sparse=False, n_values= len(labels[i]))
    feature = onehot_encoder.fit_transform(feature)
    cats.append(feature)
```

```python
encoded_cats = np.column_stack(cats)
```

```python
dataset_encoded = np.concatenate((encoded_cats, dataset.iloc[:,116:].values),axis=1)
```

```python
del cats
del feature
del dataset
del encoded_cats
```

```python
dataset_encoded.shape
```

```python
r, c = dataset_encoded.shape
```

```python
i_cols = []
for i in range(0, c-1):
    i_cols.append(i)
```

```python
X = dataset_encoded[:,0:(c-1)]
Y = dataset_encoded[:,(c-1)]
```

```python
val_size = 0.1
seed = 0
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=val_size, random_state=seed)
```

```python
X_all =[]
comb = []
mae =[]
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
n='All'
```

```python
X_all.append([n,i_cols])
```

# Linear Regression (Linear Algorithm)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
algo = 'LR'
```

```python
for name, i_cols_list in X_all:
    model.fit(X_train[:,i_cols_list], Y_train)
    result = mean_absolute_error(np.expm1(Y_val), np.expm1(model.predict(X_val[:,i_cols_list])))
    mae.append(result)
    print result
comb.append(algo)
```

# Ridge Regression (Linear Algorithm)

```python
from sklearn.linear_model import Ridge
```

```python
a_list = np.array([0.0001,0.001,0.01,0.1,1,10,100,1000])
for alpha in a_list:
    model = Ridge(alpha = alpha, random_state=seed)
    algo = 'Ridge'
    for name, i_cols_list in X_all:
        model.fit(X_train[:,i_cols_list],Y_train)
        result = mean_absolute_error(np.expm1(Y_val), np.expm1(model.predict(X_val[:,i_cols_list])))
        mae.append(result)
        print result
    comb.append(algo + " %s" % alpha )
    
```

```python
#The larger the alpha the better the MAE - the higher the regularization the better
```

# LASSO Regression (Linear Algorithm)

```python
from sklearn.linear_model import Lasso
```

```python
a_list = np.array([0.0001,0.001,0.01,0.1,1,10,100,1000])
for alpha in a_list:
    model = Lasso(alpha = alpha, random_state=seed)
    algo = 'Lasso'
    for name, i_cols_list in X_all:
        model.fit(X_train[:,i_cols_list],Y_train)
        result = mean_absolute_error(np.expm1(Y_val), np.expm1(model.predict(X_val[:,i_cols_list])))
        mae.append(result)
        print result
    comb.append(algo + " %s" % alpha )
```

```python
#unlike in ridge, here the lower the value of alpha the better  - Lasso takes the sum of the absolute value of the magnitude or coefficient
#while ridge takes the sum of the square of the coefficients
```

# Elastic Net Regression (Linear Algorithm)

```python
from sklearn.linear_model import ElasticNet
```

```python
a_list = np.array([0.0001,0.001,0.01,0.1,1,10,100,1000])
for alpha in a_list:
    model = ElasticNet(alpha = alpha, random_state=seed)
    algo = 'ElasticNet'
    for name, i_cols_list in X_all:
        model.fit(X_train[:,i_cols_list],Y_train)
        result = mean_absolute_error(np.expm1(Y_val), np.expm1(model.predict(X_val[:,i_cols_list])))
        mae.append(result)
        print result
    comb.append(algo + " %s" % alpha )
```

```python
#the best MAE is with an alpha of 0.0001
```
